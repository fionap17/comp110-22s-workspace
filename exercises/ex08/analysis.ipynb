Next, ... (you take it from here and add additional code and markdown cells to read in the CSV file and process it as needed)

```python
SURVEY_DATA_CSV_FILE_PATH: str = "../../data/survey.csv"
data_rows: list[dict[str, str]] = read_csv_rows(SURVEY_DATA_CSV_FILE_PATH)
total_table: dict[str, list[str]] = columnar(data_rows)
```

### Comparing/looking at just answers for prior experience and synchronous lecture performance in a table format.

```python
comparison: list[str] = ["prior_exp", "sync_perf"]
selected_comparison: dict[str, list[str]] = select(total_table, comparison)
tabulate(head(selected_comparison, 620), selected_comparison.keys(), "html")
```

### Counting the number of students that identify with each amount of prior coding experience.

```python
experience: list[str] = total_table["prior_exp"]
count(experience)
```

### Counting the number of students that think their performance in this course would improve if lecture became synchronous for both days.

```python
improvement: list[str] = total_table["sync_perf"]
count(improvement)
```

```python
answers: list[str] = ["5", "6", "7"]
table: dict[str, int] = count(improvement)
improved(table, answers)
```

### Discovering the fraction of students that have none to less than one month of prior coding experience that thinks they will perform better in the course if lecture became synchronous for both days. 

```python
correlate(total_table, "prior_exp", "sync_perf")
```

### Visual for fraction of students with none to less than one month of prior coding experience that thinks synchronous lectures will improve their performance in class versus the fraction that does not think synchronous lectures will improve their performance.

```python
import matplotlib.pyplot as plt

performance: dict[str, int] = improved(table, answers)
plt.bar(performance.keys(), performance.values())
plt.title("Students with little prior coding experience and their thoughts on synchronous lectures on their performance.")
plt.xlabel('Performance in Class')
plt.ylabel("Number of Students")
plt.show()
```
